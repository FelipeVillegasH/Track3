```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('/content/drive/MyDrive/CodingDojo/Track 3/Core/Week 4 /School_Details.csv')
df.head()
```

```python
# instantiating the engine
engine = create_engine('sqlite:///schools.db', echo = True)
connection = engine.connect()
```

```python
# Dataframe in SQL format: 
sql = 'school_detail'
df.to_sql(sql, connection, if_exists = 'fail')
```

The following code serves the purpose to select certain variables to recreate dataframes in SQL

```python
sql = """
SELECT EstablishmentName, SchoolLevel, Website
FROM school_details
"""

df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName, "DistrictHQDistance (in Km)"
FROM school_details
WHERE "DistrictHQDistance (in Km)" > 100
"""

df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)"
FROM school_details
WHERE SchoolLevel = 'JHS'
"""

df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)"
FROM school_details
WHERE EstablishmentName LIKE '%JUNIOR HIGH%'
"""

df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName, LocatedInRuralOrUrban
FROM school_details
WHERE LocatedInRuralOrUrban LIKE 'Urban%'
ORDER BY EstablishmentName DESC
"""

df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName AS 'Nombre', LocatedInRuralOrUrban
FROM school_details
WHERE LocatedInRuralOrUrban LIKE 'Urban%'
ORDER BY EstablishmentName DESC
"""

df = pd.read_sql_query(sql, engine)
df.head()
```
