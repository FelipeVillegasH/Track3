```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
X = dftrain.drop(columns = ['Survived','Name','Ticket','Cabin'])
y = dftrain['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3)
```

```python
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
print('Puntaje de Entrenamiento:', xgb.score(X_train, y_train))
print('Puntaje de Testeo:', xgb.score(X_test, y_test))
```

```python
Testmodel = dftest.drop(columns = ['Name','Ticket','Cabin'])
```

```python
predictions = xgb.predict(Testmodel)
predictions
```

```python
Survivedarray = pd.DataFrame({'PassengerId' : dftest['PassengerId'], 'Survived' : predictions})
Survivedarray
```
